```python
df= pd.read_parquet('/gpfs/data1/vclgp/xiongl/ProjectIS2CalVal/data/is2_calval_filter_add_snr_v20240306.parquet')
```

```python
# filter is very important
# strong beams 
tmp  = df[df['Strong_flag']]
# # remove leaf off shots 
tmp = tmp[tmp['leaf_off_p1'] < 1]
tmp = tmp[tmp['leaf_off_p2'] < 1]
# terrain uncertainty 
tmp  = tmp[tmp['land_segments/terrain/h_te_uncertainty'] < 10000]
# h canopy uncertainty 
tmp = tmp[tmp['land_segments/canopy/h_canopy_uncertainty'] < 10000]
# canopy photons
tmp = tmp[tmp['land_segments/canopy/n_ca_photons'] < 140]
# canopy cover > 0
tmp = tmp[tmp['treecover_value'] > 0]
# canopy big error
tmp = tmp[tmp['land_segments/canopy/h_canopy_20m'] < 120]
# als big error
tmp = tmp[tmp['als_rh98'] < 120]
# simulation big error
tmp = tmp[tmp['h_canopy_98'] > 0]
tmp = tmp[tmp['h_canopy_98'] < 120]
# als z ground big error
tmp = tmp[tmp['z_ground'] > -1000]
# ['als_rh0']
tmp = tmp[tmp['als_rh0'] > -10]
tmp = tmp[(tmp['als_rh1'] - tmp['als_rh0']) < 100]

# big dif fitler 
tmp = tmp[abs(tmp['land_segments/canopy/h_canopy_20m'] - tmp['h_canopy_98']) <20 ]
tmp = tmp[abs(tmp['als_rh98'] - tmp['h_canopy_98']) < 5 ]
#tmp = tmp[abs(tmp['land_segments/canopy/h_canopy_20m'] - tmp['als_rh98']) <20 ]
# remove unwant cols
columns_to_drop = ['system:index', 'fid']
tmp.drop(columns=columns_to_drop, inplace=True)
tmp.rename(columns={'Strong_flag': 'strong_flag'}, inplace=True)

# drop bad sites
drop_list = ['neon_blan2022', 'neon_clbj2022', 'neon_moab2021', 'tern_robson_whole','embrapa_brazil_2020_hum_a01', 'costarica_laselva2019', 'embrapa_brazil_2020_duc_a01']
filtered_tmp = tmp.groupby('site_name').filter(lambda x: x.name not in drop_list)
tmp = filtered_tmp.reset_index(drop=True)

# add disturbance flag to remove disturbed shots

# possible improvements 
# terrain_flg Terrain flag quality check to indicate a deviation
# subset_can_flag Quality flag indicating the canopy photons populating the 100m segment statistics
# subset_te_flag
# qa_granule_pass_fail Flag indicating granule quality
# layer_flag Flag Meanings: ['likely_clear', 'likely_cloudy']
# msw_flag  https://www.sciencedirect.com/science/article/pii/S0034425723001219 cloud contamination
# segment_snowcover ['ice_free_water', 'snow_free_land', 'snow','ice']
```

```python
# get loop up table 
# layer  PFT 4 layers 
# EBT  2 
# DBT 4
# Needleleaf Trees[1,3] , very few shots in DBT[4]
# GSW (grasses, shrubs, and woodlands) [5,6]
# slope , cover 
# NT layer
def layer_function(row):
    layer_name = 'nan'
    if row['pft_value'] == 1 or row['pft_value'] == 3:
        layer_name = 'nt'
    if row['pft_value'] == 2:
        layer_name = 'ebt'
    if row['pft_value'] == 4:
        layer_name = 'dbt'
    if row['pft_value'] == 5 or row['pft_value'] == 6:
        layer_name = 'gsw'
    return layer_name
tmp['strata'] = tmp.apply(lambda row: layer_function(row), axis=1)
```

```python
# slope cover 5 degree bins
bins = list(range(0, 101, 5))  # From 0 to 100 with an interval of 5
tmp['slope_category'] = pd.cut(tmp['nasadem_slope'], bins, right = False) # [0.0, 5.0)  
tmp['cover_category'] = pd.cut(tmp['treecover_value'], bins, right = False) # [0.0, 5.0) # max = 91
# # slope layers 
# bins = [0, 5, 10, 15, 20, 100]  # Define the bin edges
# labels = ['0-5', '5-10', '10-15', '15-20', '20-100'] #7
# tmp['slope_category'] = pd.cut(tmp['nasadem_slope'], bins, labels = labels)
# canopy cover layers 
# bins = [0, 20, 40, 60, 70, 100]  # Define the bin edges
# labels = ['0-20', '20-40', '40-60', '60-70', '70-100']
# tmp['cover_category'] = pd.cut(tmp['treecover_value'], bins, labels = labels)
```

```python
res = []
# loop through every layer 
for l in tmp['strata'].unique():
    if l == 'nan': continue
    l_data = tmp[tmp['strata'] == l]
    for s in tmp['slope_category'].unique().dropna():
        s_data = l_data[l_data['slope_category'] == s].copy()
######## each layer diferent cover category 
        # if l == 'ebt':
        #     bins = [0, 20, 40, 60, 70, 100]  # Define the bin edges
        #     labels = ['0-20', '20-40', '40-60', '60-70', '70-100']
        #     s_data['cover_category'] = pd.cut(s_data['treecover_value'], bins, labels = labels)
        # if l == 'dbt' or  l == 'nt':
        #     bins = [0, 20, 40, 60,100]  # Define the bin edges
        #     labels = ['0-20', '20-40', '40-60', '60-100']
        #     s_data['cover_category'] = pd.cut(s_data['treecover_value'], bins, labels = labels)
        # if l == 'gsw':
        #     bins = [0, 10, 20,40, 100]  # Define the bin edges
        #     labels = ['0-10', '10-20', '20-40', '40-100']
        #     s_data['cover_category'] = pd.cut(s_data['treecover_value'], bins, labels = labels)
        for c in s_data['cover_category'].unique().dropna():
                c_data = s_data[s_data['cover_category'] == c]
                x = c_data['h_canopy_98']
                y = c_data['land_segments/canopy/h_canopy_20m']
                rmse = np.sqrt(np.mean((x - y) ** 2))
                res.append([l,c,s,rmse, len(c_data)])
                # if len(c_data) < 200:
                #     plt.scatter(x, y)
                #     plt.show()
                
data = pd.DataFrame(res, columns=['strata','cover_category', 'slope_category', 'measure_var', 'n'])
```

```python
def get_s_c_bins(data, strata):
    res = []
    for s in data['slope_category'].unique().dropna():
        s_data = data[data['slope_category'] == s]
        for c in s_data['cover_category'].unique().dropna():
                c_data = s_data[s_data['cover_category'] == c]
                res.append([strata, c,s,len(c_data)])             
    data_calval = pd.DataFrame(res, columns=['strata', 'cover_category', 'slope_category', 'n'])
    return data_calval
```

```python
bins_nt = get_s_c_bins(df_nt, 'nt')
bins_ebt = get_s_c_bins(df_ebt, 'ebt')
bins_dbt = get_s_c_bins(df_dbt, 'dbt')
bins_gsw = get_s_c_bins(df_gsw, 'gsw')
```

```python
# merge
bins_df_global = pd.concat([bins_nt, bins_ebt, bins_dbt, bins_gsw], ignore_index=True)
bins_df_global['cover_mid'] = bins_df_global['cover_category'].apply(lambda x: x.mid)
bins_df_global['slope_mid'] = bins_df_global['slope_category'].apply(lambda x: x.mid)
```

```python
# global vs calval comparison 
# nt 
# Create a figure and two subplots
def plot_hist(all, calval, name):
        fig, axs = plt.subplots(2, 2, figsize=(8, 5))
        
        # Plot for subplot 1
        axs[0,0].hist(all.cover, bins=20, color='skyblue', edgecolor='black')
        axs[0,0].set_title(f'Global_{name}')
        axs[0,0].set_xlabel('Canopy cover')
        axs[0,0].set_ylabel('N')
        # Plot for subplot 2
        axs[0,1].hist(calval.treecover_value, bins=20, color='red', edgecolor='black')
        axs[0,1].set_title(f'Calval_{name}')
        axs[0,1].set_xlabel('Canopy cover')
        axs[0,1].set_ylabel('N')
        
        axs[1,0].hist(all.slope, bins=20, color='skyblue', edgecolor='black')
        axs[1,0].set_title(f'Global_{name}')
        axs[1,0].set_xlabel('Slope')
        axs[1,0].set_ylabel('N')
        # Plot for subplot 2
        axs[1,1].hist(calval.nasadem_slope, bins=20, color='red', edgecolor='black')
        axs[1,1].set_title(f'Calval_{name}')
        axs[1,1].set_xlabel('Slope')
        axs[1,1].set_ylabel('N')
        # Adjust layout
        plt.tight_layout()
        plt.savefig(f'../tmp_test/histogram_{name}.png', bbox_inches='tight', pad_inches=0)
        # Show plot
        plt.show()
```

```python
plot_hist(df_nt, tmp_nt, 'nt')
plot_hist(df_ebt, tmp_ebt, 'ebt')
plot_hist(df_dbt, tmp_dbt, 'dbt')
plot_hist(df_gsw, tmp_gsw, 'gsw')
```

```python
# bins
def layer_function(row):
    layer_name = 'nan'
    if row['pft_value'] == 1 or row['pft_value'] == 3:
        layer_name = 'nt'
    if row['pft_value'] == 2:
        layer_name = 'ebt'
    if row['pft_value'] == 4:
        layer_name = 'dbt'
    if row['pft_value'] == 5 or row['pft_value'] == 6:
        layer_name = 'gsw'
    return layer_name
df_calval['strata'] = df_calval.apply(lambda row: layer_function(row), axis=1)
```

```python
# slope 
bins = [0, 5, 10, 15, 20, 100]
df_calval['slope_category'] = pd.cut(df_calval['nasadem_slope'], bins=bins, right=False)
```

```python
# cover
res = []
# loop through every layer 
for l in tmp['strata'].unique():
    if l == 'nan': continue
    l_data = df_calval[df_calval['strata'] == l]
    for s in df_calval['slope_category'].unique().dropna():
        s_data = l_data[l_data['slope_category'] == s].copy()
######## each layer diferent cover category 
        if l == 'ebt': #  0-10, 10-15, 15-20, ....,65-70, 70-75,75-80, >80
            bins = [0, 10, 15,20,25,30,35, 40,45,50,55, 60,65, 70,75,80,100]  # Define the bin edges
            s_data['cover_category'] = pd.cut(s_data['treecover_value'], bins, right=False)
        if l == 'dbt' or  l == 'nt': # cover: 0-10, 10-15, 15-20, ....,65-70, >70
            bins = [0,10,15,20,25,30,35,40,45,50,55,60,65,70,100]  # Define the bin edges
            s_data['cover_category'] = pd.cut(s_data['treecover_value'], bins, right=False)
        if l == 'gsw': # 0-5, 5-10, 10-15, 15-20, ..., 35-40, >40
            bins = [0, 5 ,10, 15, 20, 25, 30,35, 40, 100]  # Define the bin edges
            s_data['cover_category'] = pd.cut(s_data['treecover_value'], bins, right=False)
        for c in s_data['cover_category'].unique().dropna():
                c_data = s_data[s_data['cover_category'] == c]
                x = c_data['h_canopy_98']
                y = c_data['land_segments/canopy/h_canopy_20m']
                rmse = np.sqrt(np.mean((x - y) ** 2))
                res.append([l,c,s,rmse, len(c_data)])
                if len(c_data) < 100:
                    plt.scatter(x, y)
                    plt.xlim(0,50)
                    plt.ylim(0,50)
                    plt.show()             
data = pd.DataFrame(res, columns=['strata','cover_category', 'slope_category', 'measure_var', 'n'])
```

```python
# get ease tiles and gridding 
def get_tile_id(x, y, tilesize=72):
    ease2_origin = -17367530.445161499083042, 7314540.830638599582016
    ease2_nbins = int(34704 / tilesize), int(14616 / tilesize)
    ease2_binsize = 1000.895023349556141*tilesize, 1000.895023349562052*tilesize 
    #transformer = Transformer.from_crs('epsg:4326', 'epsg:6933', always_xy=True)
    #x,y = transformer.transform(longitude, latitude)
    xidx = int( (x - ease2_origin[0]) / ease2_binsize[0]) + 1
    yidx = int( (ease2_origin[1] - y) / ease2_binsize[1]) + 1
    return xidx, yidx
```

```python
def get_tile_bbox(x, y, tilesize=72):
    ease2_origin = -17367530.445161499083042, 7314540.830638599582016
    ease2_nbins = int(34704 / tilesize), int(14616 / tilesize)
    ease2_binsize = 1000.895023349556141*tilesize, 1000.895023349562052*tilesize
    
    xmin = ease2_origin[0] + (x - 1) * ease2_binsize[0]
    xmax = xmin + ease2_binsize[0]
    ymax = ease2_origin[1] - (y - 1) * ease2_binsize[1]
    ymin = ymax - ease2_binsize[1]
    return xmin,xmax,ymin,ymax
```

```python
# Create an empty DataFrame
res = []
# Define the range of x and y coordinates
min_x = -17367500
max_x = 17367500
min_y = -7314500
max_y = 7314500
# Iterate over x and y ranges
for x in range(min_x, max_x, 5000):
    for y in range(min_y, max_y, 5000):
        # Get tile indices for the current coordinate pair
        xidx, yidx = get_tile_id(x, y)
        # Append tile indices to the list as a dictionary
        res.append({'xidx': xidx, 'yidx': yidx})
res = pd.DataFrame(res)
all_tiles = res.drop_duplicates(subset=['xidx', 'yidx']) # X[1-482] , Y[1-203]
print(res)
```
