```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the two Poisson distributions
lambda_vals = [0.5, 1]
size = 1000  # Number of samples

# Generate Poisson distribution data
data_05 = np.random.poisson(lam=lambda_vals[0], size=size)
data_01 = np.random.poisson(lam=lambda_vals[1], size=size)

# Create the plot
plt.figure(figsize=(8, 6))
# Plot both distributions
plt.hist(data_05, bins=30, color='green', alpha=0.6, label='λ=0.5')
plt.hist(data_01, bins=30, color='orange', alpha=0.6, label='λ=1')
# Add titles and labels
plt.title('Poisson Distributions (λ=0.5 and λ=0.1)', fontsize=16)
plt.xlabel('Number of Events', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xlim(0,20)
plt.legend()
# Show the plot
plt.show()
```
