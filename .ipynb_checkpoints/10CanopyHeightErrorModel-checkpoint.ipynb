```python
# add leaf off flag
# Define a function to be applied to each row
def leaf_off_p1(row):
    flag = 0
    # value is not valid.
    if row['greeness_max_p1'] > 366 or row['greeness_dec_p1'] > 366: return flag
    # delta time to date seconds since 2018-01-01
    delta = row['land_segments/delta_time']
    result_datetime = datetime(2018, 1, 1) + timedelta(seconds=delta)
    days = (result_datetime - datetime(result_datetime.year, 1, 1)).days + 1
    if row['land_segments/latitude_20m'] >= 0:
        if days < row['greeness_max_p1'] or days >  row['greeness_dec_p1']: flag = 1
    else:
        if days < row['greeness_max_p1'] and  days >  row['greeness_dec_p1']: flag = 1
    return flag
def leaf_off_p2(row):
    flag = 0
    # value is not valid.
    if row['greeness_max_p2'] > 366 or row['greeness_dec_p2'] > 366: return flag
    # delta time to date seconds since 2018-01-01
    delta = row['land_segments/delta_time']
    result_datetime = datetime(2018, 1, 1) + timedelta(seconds=delta)
    days = (result_datetime - datetime(result_datetime.year, 1, 1)).days + 1
    if row['land_segments/latitude_20m'] >= 0:
        if days < row['greeness_max_p2'] or days >  row['greeness_dec_p2']: flag = 1
    else:
        if days < row['greeness_max_p2'] and  days >  row['greeness_dec_p2']: flag = 1
    return flag
# south month 12 -2 next year 
# north  month 6-8 
```

```python
# Apply the function to each row and create a new column with the result
df['leaf_off_p1'] = df.apply(leaf_off_p1, axis=1)
df['leaf_off_p2'] = df.apply(leaf_off_p2, axis=1)
```

```python
# x y plot  for all beams, strong beams, and weak beams.
import numpy as np
import matplotlib.pyplot as ax
def plot_xy (ax, x, y, vmax=400, set_y_label = None , sub = '(a) all beams'):
    # x is reference
    # get parameter
    n = len(x)
    n_s = "{:,.0f}".format(n)
    r2 = np.corrcoef(y, x)[0, 1]**2
    r2_s= "{:.2f}".format(round(r2,2))
    bias = np.mean(y-x)
    bias_s = "{:.2f}".format(round(bias,2))
    mae = np.mean(abs(y-x))
    mae_s = "{:.2f}".format(round(mae,2))
    rmse = np.sqrt(np.mean((x-y)**2))
    rmse_s = "{:.2f}".format(rmse.round(2))
    p_rmse = rmse/np.mean(x)*100
    p_rmse_s = "{:.0f}".format(p_rmse.round(0))
    # Create 2D density plot
    # Create 2D density plot
    h = ax.hist2d(x, y, bins=100, cmap='OrRd', vmin=0, vmax=vmax) # OrRd
    #ax.colorbar(label='Density')
    ax.set_xlabel('ALS p98 [m]')
    if set_y_label is not None:
        ax.set_ylabel('ICESat-2 canopy height [m]')     
    ax.set_xlim(0, 60)
    ax.set_ylim(0, 60)
    # Insert text
    ax.text(0.05, 0.50, f'N = {n_s}\n$R^2$ = {r2_s}\nBias = {bias_s}\nMAE = {mae_s}\nRMSE = {rmse_s}\n%RMSE = {p_rmse_s}', transform=ax.transAxes)
    ax.text(0.45, 0.1, f'{sub}', transform=ax.transAxes)
    ax.plot([0, 100], [0, 100], color='black', linestyle='--')

    #ax.title('2D Density Plot')
    #plt.show()
```

```python
def plot_slope_cover_bin(tmp_go, name):
        tmp_go=tmp_go[tmp_go['treecover_value'] >0]
        tmp_go=tmp_go[tmp_go['treecover_value'] <100]
        res = []
        for c in range(25, 980, 50):
            c_bin_l = c/10-2.5
            c_bin_r = c/10+2.5
            for s in range(25, 980, 50):
                s_bin_l = s/10-2.5
                s_bin_r = s/10+2.5
                filtered_df = tmp_go[(tmp_go['treecover_value'] >= c_bin_l) & (tmp_go['treecover_value'] < c_bin_r) &
                                    (tmp_go['nasadem_slope'] >= s_bin_l) & (tmp_go['nasadem_slope'] < s_bin_r)]
                x = filtered_df['h_canopy_98']
                y = filtered_df['land_segments/canopy/h_canopy_20m']
                rmse = np.sqrt(np.mean((x - y) ** 2))
                res.append([c/10,s/10,rmse, len(x)])
        data = pd.DataFrame(res, columns=['x_cover', 'y_slope', 'z_rmse', 'N'])
        x= np.linspace(2.5, 97.5, 20)
        y =np.linspace(2.5, 97.5, 20)
        X,Y = np.meshgrid(x, y)
        Z = interpolate.griddata((data['x_cover'], data['y_slope']), data['z_rmse'], (X,Y), method='linear')
        plt.figure(figsize=(8, 8))
        gs = gridspec.GridSpec(2, 2,  width_ratios=(4, 1), height_ratios=(1, 4),
                              left=0.1, right=0.9, bottom=0.1, top=0.9,
                              wspace=0.05, hspace=0.05)
        ax_main = plt.subplot(gs[1, 0])
        #ax_test = plt.subplot(gs[0, 1])
        ax_xDist = plt.subplot(gs[0, 0], sharex=ax_main)
        ax_yDist = plt.subplot(gs[1, 1], sharey=ax_main)
        mesh = ax_main.pcolormesh(X,Y,Z,cmap='OrRd',vmax=12)
        ax_main.text(0.05, 0.95, f'{name}', horizontalalignment='left',
                 verticalalignment='center', transform=ax_main.transAxes)
        ax_main.xaxis.set_minor_locator(ticker.MultipleLocator(5))
        ax_main.yaxis.set_minor_locator(ticker.MultipleLocator(5))
        colorbar_position = plt.axes([0.2, 0.65, 0.4, 0.02])  # [left, bottom, width, height]
        cbar = plt.colorbar(mesh, ax=ax_main, orientation = 'horizontal', shrink=0.6, cax = colorbar_position )
        cbar.set_ticks(np.linspace(0, 12, num=7))
        ax_main.set_xlim(0, 99)  
        ax_main.set_ylim(0, 99) 
        # Add labels and title
        bins = np.arange(0, 101, 2)
        ax_main.set(xlabel='Canopy cover [%]', ylabel='Terrain Slope [\u00b0]')
        ax_xDist.hist(tmp_go['treecover_value'],bins=bins,align='mid', color='gray')
        ax_xDist.tick_params(axis="x", labelbottom=False)
        ax_yDist.hist(tmp_go['nasadem_slope'],bins=bins,orientation='horizontal',align='mid', color='gray')
        ax_yDist.tick_params(axis="y", labelleft=False)        
        plt.savefig('../report/is2error/rmse_is2_'+name+'.png')
        plt.close()
        #plt.show()
```

```python
# glm model ; Generalized Linear Models
#!pip install statsmodels --upgrade
import statsmodels.api as sm
model = sm.GLM(y_train, X_train, family=sm.families.Gaussian())
model_result = model.fit()
# Print the summary of the model
print(model_result.summary())
```
