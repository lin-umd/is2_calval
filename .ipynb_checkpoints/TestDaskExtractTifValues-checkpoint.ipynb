```python
import pandas as pd
import dask.dataframe as dd
```

```python
df = pd.read_parquet('/gpfs/data1/vclgp/xiongl/ProjectIS2CalVal/report/is2_calval_stat_v20240129.parquet')
df_sample = df[:100]
#ddf = dd.from_pandas(df[['land_segments/longitude_20m','land_segments/latitude_20m' ]], npartitions=5)
```

```python
ddf = dd.from_pandas(df[['land_segments/longitude_20m','land_segments/latitude_20m','land_segments/delta_time']], npartitions=1000)
ddf
```

```python
import os
import rasterio
tif_max = '../data/phenology/GEDI_ANCI_PHENO_Date_Onset_Greenness_Maximum_Phase2_v5_epsg6933_1km.tif'
tif_dec = '../data/phenology/GEDI_ANCI_PHENO_Date_Onset_Greenness_Decrease_Phase2_v5_epsg6933_1km.tif'
gm = rasterio.open(tif_max)
gd = rasterio.open(tif_dec)
band_m = gm.read(1)
band_d = gd.read(1)
```

```python
from pyproj import Transformer
transformer = Transformer.from_crs("EPSG:4326", "EPSG:6933", always_xy=True)
x, y = transformer.transform(-110, 40)
row, col = gm.index(x, y)
band_m[row, col]
band_d[row, col]
```

```python

def extract_gm(row):
    lon = row['land_segments/longitude_20m']
    lat = row['land_segments/latitude_20m']
    delta=row['land_segments/delta_time']
    x, y = transformer.transform(lon, lat)
    row_gm, col_gm = gm.index(x, y)
    row_gd, col_gd = gd.index(x, y)
    data = [lon, lat, delta, band_m[row_gm, col_gm],band_d[row_gd, col_gd]]
    names = ['land_segments/longitude_20m', 'land_segments/latitude_20m','land_segments/delta_time', 'greeness_max_p2', 'greeness_dec_p2']
    return pd.Series(data,index=names)
    # DataFrame can have multiple columns, while a Series can only have one
```

```python
# test
for index, row in df_sample[:2].iterrows():
    print(extract_gm(row))
```

```python
meta_df = {'land_segments/longitude_20m':'float64',
'land_segments/latitude_20m':'float64',
'land_segments/delta_time': 'float64',
'greeness_max_p2': 'uint16',
'greeness_dec_p2': 'uint16'}
res = ddf.map_partitions(lambda partition: partition.apply(lambda row: extract_gm(row), axis=1), meta=meta_df)
```

```python
greeness = res.compute()
```

```python
greeness.to_parquet('../result/greeness_calval_p2_02282024.parquet')
```
