```python
from dask.distributed import Client, progress
import dask.dataframe as dd
import dask
import sys
import os
import shutil
import geopandas as gpd
from shapely.geometry import Polygon
import icepyx as ipx
import argparse
import subprocess
def order_vertices_counter_clockwise(polygon):
    if not polygon.is_empty and polygon.exterior.is_ccw:
        return polygon
    else:
        return Polygon(polygon.exterior.coords[::-1])    
```

```python
import geopandas as gpd
@dask.delayed
def get_atl03_photon(f_in_polygon, name, index):
    folder_out = '/gpfs/data1/vclgp/xiongl/ProjectIS2CalVal/data/atl03_data/'+ name + '_'+ str(index)
    if os.path.exists(folder_out): return
    short_name = 'ATL03'
    date_range = ['2018-01-01','2023-12-31']
    file_list = []
    # Access the geometry of the first polygon
    polygon_geometry = f_in_polygon
    polygon_geometry = order_vertices_counter_clockwise(polygon_geometry)
    lon_lat_pairs = list(polygon_geometry.exterior.coords)
    region_a = ipx.Query(short_name, lon_lat_pairs, date_range)
    region_a.avail_granules()
    mylist = region_a.granules.avail
    if (len(mylist) == 0): return
    region_a.order_granules()
    # data/atl03_data/atl03_data_csir_agincourt
    region_a.download_granules(folder_out)
```

```python
# idea 1degree cell to intersect 
import geopandas as gpd
from shapely.geometry import box
# Assuming gdf is your GeoDataFrame
# Replace 'geometry' with the actual name of the geometry column
# Create a regular grid of 1-degree cells
lon_min, lon_max = -180, 180
lat_min, lat_max = -90, 90
cell_size = 1
grid_cells = []
for lon in range(lon_min, lon_max, cell_size):
    for lat in range(lat_min, lat_max, cell_size):
        cell = box(lon, lat, lon + cell_size, lat + cell_size)
        grid_cells.append(cell)

# Create a GeoDataFrame for the grid cells
grid_gdf = gpd.GeoDataFrame(geometry=grid_cells, crs=gdf.crs)

# Use geopandas.clip to find which geometries intersect with each cell
#intersections = gpd.clip(grid_gdf, gdf)
# Display the resulting GeoDataFrame
#print(intersections)
```

```python
def plot_photon(beam='gt1l', file=files[18]):
        with h5py.File(file,'r') as f:
            if (beam not in list(f.keys())): return
            if ('heights' not in list(f[beam].keys())): return
            beam_lat = f[f'/{beam}/heights/lat_ph'][:]
            beam_lon = f[f'/{beam}/heights/lon_ph'][:]
            beam_height = f[f'/{beam}/heights/h_ph'][:]
            beam_dist_ph = f[f'/{beam}/heights/dist_ph_along'][:]
            beam_signal_conf_ph = f[f'/{beam}/heights/signal_conf_ph'][:][:, 0]
            f.close()
        geometry = gpd.points_from_xy(beam_lon, beam_lat)
        df = {'Latitude': beam_lat, 'Longitude': beam_lon, 
              'Photon_Height': beam_height,'Along_track_distance':beam_dist_ph,
             'signal_conf_ph': beam_signal_conf_ph}
        gdf = gpd.GeoDataFrame(df,geometry=geometry, crs='EPSG:4326')
        
        # 0=noise. 1=added to allow for buffer but algorithm
        # classifies as background 2=low 3=med 4=high)
        gdf = gdf[(gdf['signal_conf_ph'] == 4)| (gdf['signal_conf_ph'] == 3) | (gdf['signal_conf_ph'] == 2)]
        if len(gdf) < 100: return
        # View the resulting geopandas dataframe
        #print(gdf.head())
        import matplotlib.pyplot as plt
        fig, ax = plt.subplots(figsize=(10, 3))
        #gdf.Photon_Height.plot(ax=ax, ls='', marker='.', ms=1)
        # Create a line plot
        plt.scatter(gdf.Latitude, gdf.Photon_Height, marker='.', color='b', label='Data Points', s=1)
        ax.set_xlabel('Latitude (degree)', fontsize=12);
        ax.set_ylabel('Photon Height (m)', fontsize=12)
        ax.set_title(beam+'_'+os.path.basename(file), fontsize=14)
        ax.tick_params(axis='both', which='major', labelsize=12)
        # Set x-axis limits
        #ax.set_xlim(-24.84,-24.785)
    #### zoom
        #ax.set_xlim(-24.82,-24.815) 
        #ax.set_ylim(570,600) 
        plt.show()
```

```python
#%matplotlib inline
for f in files[:2]:
    #if 'processed_ATL03_20190502040024_05170314_006_02.h' in f:
            for b in ['gt1l','gt1r','gt2l','gt2r','gt3l','gt3r']:
                plot_photon(beam=b, file=f)
#gdf.to_parquet('test_data/test.parquet')
```

```python
# read to table 
# Specify the path to your text file
file_path = 'check_result.txt'
import json
df = pd.DataFrame()
# Open the file in read mode
with open(file_path, 'r') as file:
    # Read the file line by line
    index = 0 
    for line in file:
        if 'Number of available granules' in line:
            # Convert the string to a dictionary
            data_dict = json.loads(line.strip().replace("'", "\""))
            # Extract the value associated with the key 'Number of available granules'
            number = data_dict.get('Number of available granules')
            # Print the result
            df.loc[index, 'number'] = number
        if '-- Done' in line:
        # Process each line (replace this with your logic)
            #print(line.strip())  # strip() removes leading and trailing whitespaces
            name = line.strip().split(' ')[-1]
            df.loc[index, 'name'] = name
            index = index + 1
# Sort the DataFrame by the 'number' column in ascending order
df_sorted = df.sort_values(by='number', ascending=False)
```
